###  Organisation et nettoyage des données

```python
# Comparé les colonnes entre les deux DF.

def compare_columns(df1,df2):
    columns_1 = list(df1.columns) 
    columns_2 = list(df2.columns)
    same_columns=[]
    diff_columns_2=[]
    diff_columns_1=[]

    for col in columns_2:
        if col in columns_1:
            same_columns.append(col)
        else:
            diff_columns_2.append(col)
    for col in columns_1:
        if col not in columns_2:
            diff_columns_1.append(col)
    return diff_columns_1, diff_columns_2
```

```python

# Étant donné que la colonne « emplacement » contient toutes les informations de manière compacte, 
# On doit la décomposer en utilisant ast- "Abstract Syntax Trees"

import ast
df_2015['Location'] = [ast.literal_eval(str(item)) for index, item in df_2015.Location.iteritems()]
df_2015 = pd.concat([df_2015.drop(['Location'], axis=1), df_2015['Location'].apply(pd.Series)], axis=1)
df_2015['human_address'] = [ast.literal_eval(str(item)) for index, item in df_2015.human_address.iteritems()]
df_2015 = pd.concat([df_2015.drop(['human_address'], axis=1), df_2015['human_address'].apply(pd.Series)], axis=1)
df_2015.head()
```

#### Fusion des deux blocs de données

```python
df = pd.concat([df_2015, df_2016])
print ("Le dataset compte {} lignes et {} variables".format(df.shape[0], df.shape[1]))
df
```

```python
# Suppression les variables qui sont pas pertinente

df.drop(['LargestPropertyUseTypeGFA', 
           'SecondLargestPropertyUseTypeGFA',
           'SecondLargestPropertyUseType',
           'ThirdLargestPropertyUseTypeGFA',
           'ThirdLargestPropertyUseType','PropertyGFAParking','ENERGYSTARScore',
           'PropertyGFABuilding(s)','DefaultData','ComplianceStatus', 'City',
                  'TaxParcelIdentificationNumber','CouncilDistrictCode'],
         axis=1, inplace=True)
```

```python
# On évite des habitations sur "Building Type". On s’intéresse de près aux émissions des bâtiments non destinés à l’habitation.

df['BuildingType'].unique()
```

```python
# On guarde le Df relevant

df = df[~df['BuildingType'].str.contains("Multifamily")]

print("Le jeu de données compte à présent {} lignes et {} colonnes.".format(df.shape[0],df.shape[1]))
```

```python

# Suppression des variables redondantes

#NaturalGas(therms) et NaturalGas(kBtu)

#Electricity(kWh) et Electricity(kBtu),

# kbtu est une mune unité commune de mesure d'énergie utilisée pour convertir et combiner d'autres mesures d'énergie 

# telles que les kilowattheures (kWh) d'électricité.

# Nous n'avons pas besoin de guarder 'NaturalGas(therms)'et'Electricity(kWh)'


redundant_colonnes = ['NaturalGas(therms)','Electricity(kWh)']

df.drop(redundant_colonnes, axis=1, inplace=True)
```

```python
# On compare les variables qui sont générées de différentes manières et donc, supprime celles qui ne sont pas pertinentes pour notr projet.

# SiteEUI(kBtu/sf) vs SiteEUIWN(kBtu/sf)            
# SourceEUI(kBtu/sf) vs SourceEUIWN(kBtu/sf)          
# SiteEnergyUse(kBtu) vs SiteEnergyUseWN(kBtu)

# on supprime aussi les variabes avec les suffix 'WN' :"Weather Normalized" qui indique les normalisation des variables par la climat.

df.drop(['SiteEUIWN(kBtu/sf)','SourceEUIWN(kBtu/sf)','SiteEnergyUseWN(kBtu)'],
         axis=1, inplace=True)
```

## Traitement de données manquantes

```python
### # En utilisant la boucle for, on calcule les données manquantes pour les colonnes


df.isnull().sum()*100/len(df)
```

```python
# On évite les Nan plus de 80 percent

colonnes_avec_nan=df.columns[df.isnull().any()]

for column in colonnes_avec_nan: 
    
    if df[column].isnull().sum()*100.0/len(df) > 80:
          
            df.drop(column,1, inplace=True)
```

```python
df.isnull().sum()*100/len(df)
```

```python
df=df.dropna(how='any')

print(df.shape)
df.head()
```

## Traitement des Outliers

```python
# Méthode des quantiles

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
print(df.shape)
df.head(5)
```

### Exploration de données

```python
## Le moyen valeur d’émission par quartier plus ou moins proche de la moyenne globale
# le Cartier 'Central' montre des émissions moyennes plus élevées et le cartier "Balad" présentent un niveau moyen d'émissions inférieur

var=[ 'BuildingType','Neighborhood', 'LargestPropertyUseType']

#This function allows you to view the distribution 
#of CO2 emissions within the classes of a variable.
def visualisation(variable,df):
    the_mean=df["TotalGHGEmissions"].mean()
    fig=plt.figure(figsize=[18,7])
    fig.patch.set_facecolor('#E0E0E0')
    fig.patch.set_alpha(0.7)
    plt.title("Répartition des émissions de C02 par {}".format(variable),size=16)
    sns.boxplot(x=variable, y="TotalGHGEmissions", data=df,color="#cbd1db",width=0.5,showfliers=False,showmeans=True)
    plt.hlines(y=the_mean,xmin=-0.5,xmax=len(df[variable].unique())-0.5,color="#6d788b",ls="--",label="Global mean")

    plt.ylabel(" C02 émissions",size=14)
    plt.xticks(range(0,len(df[variable].unique()))
               ,df[variable].unique(),rotation=90)
    plt.legend()
    plt.grid()
    plt.show()
visualisation('Neighborhood',df)
```

```python
# L'école supérieure ou l'université affiche un niveau moyen d'émissions plus élevé
visualisation('BuildingType',df)
```

```python
# Moyenne plus élevée:senior care community, whole club/superCenter,Restaurant:Bar 

visualisation('LargestPropertyUseType',df)
```
